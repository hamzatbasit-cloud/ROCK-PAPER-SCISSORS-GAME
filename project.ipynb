```python
import tkinter as tk
import random


# GUI window setup
window = tk.Tk()
window.title("ROCK-PAPER-SCISSORS GAME")
window.geometry("400x350")
window.resizable(width=False, height=False)

lbl_title = tk.Label(
    window,
    text= "Rock Paper Scissors Game",
    font=("Arial", 10, "bold")
)
lbl_title.pack(pady=10)

# Gamer choice section
user_choice = tk.StringVar(value="")

lbl_select = tk.Label(window, text="Choose One", font=("Arial", 12, "italic"))
lbl_select.pack()

# Radio Buttons selection
radio_1 = tk.Radiobutton(window, text="Rock", variable=user_choice,value="Rock")
radio_2 = tk.Radiobutton(window, text="Paper", variable=user_choice, value="Paper")
radio_3 = tk.Radiobutton(window, text="Scissors", variable=user_choice, value="Scissors")

radio_1.pack()
radio_2.pack()
radio_3.pack()


# logical function
def start_game():
    Gamer_choice = user_choice.get()

    if Gamer_choice == "":
        lbl_result.config(text ="Please choose ROCK, PAPER OR SCISSORS")
        return None
    
    computer_choice = random.choice(["Rock","Paper","Scissors"])
    lbl_computer.config(text= f"Computer choice is:{computer_choice}")

    wins = {
        "Rock":"Scissors",
        "Paper":"Rock",
        "Scissors":"Paper"
    }
    if Gamer_choice == computer_choice:
        result = "It's a TIE"
    elif wins[Gamer_choice] == computer_choice:
        result = f"You win! {Gamer_choice} beats {computer_choice}"
    
    else:
        result = f"Computer wins! {computer_choice} beats {Gamer_choice}"    

    lbl_result.config(text=result)

def reset_game():
    user_choice.set("")
    lbl_computer.config(text="")
    lbl_result.config(text="")
  
def exit_game():
    window.destroy()

# play button
btn_play = tk.Button(
    window,
    text="PLAY",
    font=("Arial",12, "italic"),
    bg="green",
    fg="black",
    command=start_game
)
btn_play.pack(pady=12)

btn_reset = tk.Button(
    window,
    text="RESET",
    font=("Arial", 12, "italic"),
    bg="yellow",
    fg="black",
    command=reset_game
)
btn_reset.place(x=250,y=300)

btn_exit_game = tk.Button(
    window,
    text="EXIT",
    font=("Arial",12, "italic"),
    bg="red",
    fg="black",
    command=exit_game
)
btn_exit_game.place(x=320,y=300)

# result labels
lbl_computer = tk.Label(window, font=("Arial", 12, "bold"))
lbl_computer.pack()

lbl_result = tk.Label(window, font=("Arial", 12, "bold"))
lbl_result.pack(pady=15)


window.mainloop()
```
